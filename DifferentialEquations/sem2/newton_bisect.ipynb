```python
import numpy as np
import matplotlib.pyplot as plt
from math import copysign
```

$$
e x^3 - 2x^2 + x + 1 = 0
$$

```python
eps = 0.01
g = lambda x: eps * x ** 3 - 2 * x ** 2 + x + 1
dg = lambda x: 3 * eps * x ** 2 - 4 * x + 1
```

```python
def newton(f, df, *, x_0: float, eps=1e-4, max_iter=1000):
    x_old, x_new = float('inf'), x_0

    for _ in range(max_iter):
        if abs(x_old - x_new) < eps:
            return x_new
        x_old = x_new
        x_new -=  f(x_new) / df(x_new)

    return None
```

```python
def bisect(f, *, x_0: float, x_1: float, eps=1e-4, max_iters=1000):
    a, b = x_0, x_1

    for _ in range(max_iters):
        mid = (a + b) / 2
        res = f(mid)
        if abs(res) < eps:
            return mid

        if copysign(1, res) == copysign(1, f(a)):
            a = mid
        else:
            b = mid

    return None
```

```python
N_x0 = newton(g, dg, x_0=-1)
print(f"f(x) = 0;  x = {N_x0:.5f}")
N_x1 = newton(g, dg, x_0=2)
print(f"f(x) = 0;  x = {N_x1:.5f}")
N_x2 = newton(g, dg, x_0=150)
print(f"f(x) = 0;  x = {N_x2:.5f}")
```

```python
B_x0 = bisect(g, x_0=-1, x_1=0)
print(f"f(x) = 0;  x = {B_x0:.5f}")
B_x1 = bisect(g, x_0=0, x_1=2)
print(f"f(x) = 0;  x = {B_x1:.5f}")
B_x2 = bisect(g, x_0=100, x_1=250)
print(f"f(x) = 0;  x = {B_x2:.5f}")
```

```python
A_x0 = 1 + eps / 3
print(f"f(x) = 0;  x = {A_x0:.5f}")
A_x1 =  - 0.5 - eps / 8
print(f"f(x) = 0;  x = {A_x1:.5f}")
A_x2 = 2 / eps - 0.5
print(f"f(x) = 0;  x = {A_x2:.5f}")
```
