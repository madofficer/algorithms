```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_bvp
```

#### Аналитическое решение
   $$ y(x) = c_1e^x+c_2xe^x} $$

```python
a2 = lambda x: 1
a1 = lambda x: -2
a0 = lambda x: 1
dx = 1e-2
x_0 = 0
y_0 = 1
x_n = 1
y_n = np.exp(2)
n = int((x_n - x_0) / dx)
```

```python
A = lambda x: a2(x) / dx ** 2 + a1(x) / dx
B = lambda x: -2 * a2(x) / dx ** 2 - a1(x) / dx + a0(x)
C = lambda x: a2(x) / dx ** 2
```

```python
x, alpha, beta = x_0, [0], [y_0]
for i in range(1, n):
    d = B(x + dx) + C(x + dx) * alpha[-1]
    alpha.append(-A(x + dx) / d)
    beta.append(-C(x + dx) * beta[-1] / d)
    x += dx
```

```python
plt.plot(np.arange(x_0, x_n, dx), alpha, label=r'$\alpha$')
plt.plot(np.arange(x_0, x_n, dx), beta, label=r'$\beta$')
plt.legend()
plt.show()
```

```python
y = [0] * (n - 1) + [y_n]
for i in range(n - 1, 0, -1):
    y[i - 1] = (alpha[i] * y[i] + beta[i])
```

```python
space = np.arange(x_0, x_n, dx)
plt.plot(space, (1 + (np.e - 1) * space) * np.exp(space), label='exact')
plt.plot(space, y, label='approx', linestyle='--')
plt.legend()
plt.show()
```
