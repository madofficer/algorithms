```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return x ** 3
```

```python
N = 100
x = np.linspace(-np.pi, np.pi, 400)
a0 = (1 / np.pi) * np.trapz(f(x), x)
an = np.zeros(N)
bn = np.zeros(N)
```

```python
for n in range(1, N + 1):
    an[n - 1] = (1 / np.pi) * np.trapz(f(x) * np.cos(n * x), x)
    bn[n - 1] = (1 / np.pi) * np.trapz(f(x) * np.sin(n * x), x)
```

```python
f_approx = a0 / 2
for n in range(1, N + 1):
    f_approx += an[n - 1] * np.cos(n * x) + bn[n - 1] * np.sin(n * x)
```

```python
plt.plot(x, f(x), label='function')
plt.plot(x, f_approx, label='Fourier series')
plt.legend()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Fourier Series Approximation')
plt.grid(True)
plt.show()
```
